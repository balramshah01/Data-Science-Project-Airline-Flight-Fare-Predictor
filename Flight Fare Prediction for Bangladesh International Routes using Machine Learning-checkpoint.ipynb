##### time-of-day period Defining

```python
def get_part_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'
Dataset['Part_of_Day'] = Dataset['Departure_Hour'].apply(get_part_of_day)
```

## TRAIN - TEST - SPLIT

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
y_log = np.log1p(Dataset['Total Fare (BDT)'])
X = Dataset.drop(columns=['Total Fare (BDT)'])
```

```python
x_train, x_test, y_train_log, y_test_log = train_test_split(x, y_log, test_size=0.2, random_state=42)
```

### Standarlization

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
import joblib
joblib.dump(scaler, 'Airline_Flight_Fare_Scaler.joblib')
```

**-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------**

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train_scaled, y_train_log)
```

```python
# Predictions
y_pred_log = lr_model.predict(x_test_scaled)
y_pred = np.expm1(y_pred_log)
y_test = np.expm1(y_test_log)
```

####  Evaluation

```python
print("Linear Regression Results:")
print("MAE :", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R Score:", r2_score(y_test, y_pred))
```

###  Actual vs Predicted of Linear Regressor

```python
plt.figure(figsize=(15,5))
plt.scatter(y_test, y_pred, alpha=0.4, color='teal')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--', color='red')
plt.title("Linear Regression: Actual vs Predicted Fares")
plt.xlabel("Actual Fare (BDT)")
plt.ylabel("Predicted Fare (BDT)")
plt.grid(True)
plt.tight_layout()
plt.savefig('Linear_Regression_Actual_vs_Predicted_Fares.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print(x_train.shape, x_test.shape)
print(np.array_equal(x_train, x_test))  
```
